```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/honeyproduction.csv')

# Display the first few rows of the dataframe
print(df.head())

# Summary statistics of the dataframe
print(df.describe())

# Plotting the total honey production per year
total_prod_per_year = df.groupby('year')['totalprod'].sum()
plt.figure(figsize=(10, 6))
plt.plot(total_prod_per_year.index, total_prod_per_year.values, marker='o')
plt.title('Total Honey Production per Year')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.grid(True)
plt.show()

# Plotting the average price per pound of honey per year
avg_price_per_year = df.groupby('year')['priceperlb'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_price_per_year.index, avg_price_per_year.values, marker='o', color='orange')
plt.title('Average Price per Pound of Honey per Year')
plt.xlabel('Year')
plt.ylabel('Average Price per Pound')
plt.grid(True)
plt.show()

# Plotting the total number of colonies per year
total_colonies_per_year = df.groupby('year')['numcol'].sum()
plt.figure(figsize=(10, 6))
plt.plot(total_colonies_per_year.index, total_colonies_per_year.values, marker='o', color='green')
plt.title('Total Number of Colonies per Year')
plt.xlabel('Year')
plt.ylabel('Total Number of Colonies')
plt.grid(True)
plt.show()

print("Data visualizations completed.")
```
